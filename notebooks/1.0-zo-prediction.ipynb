```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
dataset_train = pd.read_csv("2020Q1Q2Q3Q4-2021Q1 - Russia - Sberbank Rossii PAO (SBER).csv")
```

```python
dataset_train = dataset_train.drop(311)
```

```python
dataset_train = dataset_train.iloc[::-1]
```

```python
dataset_train = dataset_train.reset_index()
```

```python
dataset_train = dataset_train.drop(["index"], axis=1)
```

```python
training_set = dataset_train.iloc[:250, 1:2].values
```

```python
real_stock_price = dataset_train.iloc[250:, 1:2].values.tolist()
a = []
for i in range(61):
    a.append(float(real_stock_price[i][0]))
plt.plot(a, color = 'black')
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
X_train = []
y_train = []
for i in range(7, 250):
    X_train.append(training_set_scaled[i-7:i, 0])
    y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense 
```

```python
model = Sequential()

model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50))
model.add(Dropout(0.2))

model.add(Dense(units=1))

model.compile(optimizer='adam',loss='mean_squared_error')

model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
dataset_total = dataset_train[["Price"]]
inputs = dataset_total[len(dataset_total) - len(real_stock_price) - 7:].values
```

```python
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
X_test = []
for i in range(7, 68):
    X_test.append(inputs[i-7:i, 0])
```

```python
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = model.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
plt.plot(a, color = 'black', label = 'SBER Stock Price')
plt.plot(predicted_stock_price, color = 'green', label = 'Predicted SBER Stock Price')
plt.title('SBER Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('SBER Stock Price')
plt.legend()
plt.show()
```
