```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = yf.download('MSFT','2020-01-01', '2023-09-09')
```

```python
data = data.reset_index()
```

```python
data
```

```python
training_set = data.iloc[:756, 5:6].values
```

```python
real_stock_price = data.iloc[756:, 5:6].values.tolist()
a = []
for i in range(172):
    a.append(float(real_stock_price[i][0]))
plt.plot(a, color = 'black')
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)
```

```python
X_train = []
y_train = []
for i in range(7, 756):
    X_train.append(training_set_scaled[i-7:i, 0])
    y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense 
```

```python
model = Sequential()

model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50,return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(units=50))
model.add(Dropout(0.2))

model.add(Dense(units=1))

model.compile(optimizer='adam',loss='mean_squared_error')

model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
dataset_total = data[["Adj Close"]]
inputs = dataset_total[len(dataset_total) - len(real_stock_price) - 7:].values
```

```python
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
X_test = []
for i in range(7, 178):
    X_test.append(inputs[i-7:i, 0])
```

```python
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = model.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
plt.plot(a, color = 'black', label = 'MSFT Stock Price')
plt.plot(predicted_stock_price, color = 'green', label = 'Predicted MSFT Stock Price')
plt.title('MSFT Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('MSFT Stock Price')
plt.legend()
plt.show()
```

```python
real_stock_price1 = data.iloc[756:, 5:6].values.tolist()
```

```python
b = []
for i in range(172):
    b.append(float(real_stock_price1[i][0]))
```

```python
b = pd.Series(b)
```

```python
B_MA = pd.Series((b.rolling(5, min_periods=5).mean()), name='B_MA')
sigma = b.rolling(5, min_periods=5).std()
```

```python
BU = pd.Series((B_MA + 2 * sigma), name='BU')
BL = pd.Series((B_MA - 2 * sigma), name='BL')
```

```python
BU1 = BU[4:]
BL1 = BL[4:]
BU1 = BU1.reset_index(drop=True)
BL1 = BL1.reset_index(drop=True)
```

```python
plt.plot(a, color = 'black', label = 'MSFT Stock Price')
plt.plot(BU1, color = 'orange', label = 'BU')
plt.plot(BL1, color = 'blue', label = 'BL')
plt.title('MSFT Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('MSFT Stock Price')
plt.legend()
plt.show()
```
