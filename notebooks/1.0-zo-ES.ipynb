```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = yf.download('MSFT','2020-01-01', '2023-09-09')
```

```python
data = data.reset_index()
```

```python
training_set = data.iloc[:756, 5:6]
```

```python
real_stock_price = data.iloc[756:, 5:6].values.tolist()
a = []
for i in range(172):
    a.append(float(real_stock_price[i][0]))
plt.plot(a, color = 'black')
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
model = SimpleExpSmoothing(training_set)
```

```python
fit1 = model.fit()
pred1 = fit1.forecast(172)
```

```python
pred1 = pred1.reset_index(drop=True)
```

```python
plt.plot(pred1)
plt.plot(a, color='red')
```
